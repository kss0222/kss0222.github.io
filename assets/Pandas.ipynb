```
import pandas as pd
data = {"fruits": ["apple", "orange", "banana", "strawberry", "kiwifruit"],
        "time": [1, 4, 5, 6, 3],
        "year": [2001, 2002, 2001, 2008, 2006]}
df = pd.DataFrame(data)

# drop()을 이용하여 df의 0,1행을 삭제합니다
df_1 = df.drop(range(0, 2)) #df_1 = df.drop(1) 1행만 삭제

# drop()을 이용하여 df의 열 "year"를 제거합니다
df_2 = df.drop("year", axis=1)

print(df_1)
print()
print(df_2)
```

```
import pandas as pd

index = ["Kim", "Lee", "Park", "Hong", "Han"]
columns = ["C", "Python", "Java", "Android", "C#"]
data = [[30, 45, 12, 45, 87], [65, 47, 83, 17, 58], [64, 63, 86, 57, 46,], [38, 47, 62, 91, 63], [65, 36, 85, 94, 36]]
df = pd.DataFrame(data, index=index, columns=columns)

# Python을 오름차순으로 정렬합니다
df1 = df.sort_values(by="Python", ascending=True)
print(df1)
print()

# df1의 각 요소에 5점을 더하세요
df2 = df1 + 5
print(df2)
print()

# df의 통계 정보 중에서 "mean", "max", "min"을 출력하세요
print(df2.describe().loc[["mean", "max", "min"]])
```
